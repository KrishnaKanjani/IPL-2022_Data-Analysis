# This is the data analysis of IPL 2022. The data I have used is taken from Kaggle.
## Lets Begin !

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Load and read the data
data = pd.read_csv("dataset.csv")
```

## Exploratory Data Analysis

```python
data.head(5)
```

```python
#Total Number of matches
print('Total Number of Matches:', len(data))
```

```python
#Basic Info of the data
data.info()
```

## Number of Matches Won in IPL 2022

```python
plt.figure(figsize=(9, 4))
ax = sns.countplot(x="match_winner", data=data)
ax.bar_label(ax.containers[0], fontsize=10);
plt.title("Number of Matches Won in IPL 2022")
plt.show()
```

#### Gujarat is leading the tournament by winning 12 matches.

## Matches Won by Wickets or Runs

```python
plt.figure(figsize=(4, 4))
won_by = data["won_by"].value_counts()
plt.pie(won_by, labels=won_by.index, colors=['green', 'yellow'], autopct='%1.0f%%')
plt.title('Number of Matches Won By Defending Or Chasing')
plt.show()
```

#### So, 37 matches are won while chasing the target i.e Wickets, and 37 matches are won while defending the target i.e Runs.

## Toss Decision 

```python
plt.figure(figsize=(4, 4))
toss = data["toss_decision"].value_counts()
plt.pie(toss, labels=toss.index, colors=['green', 'yellow'], autopct = '%1.0f%%')
plt.title('Toss Decision')
plt.show()
```

#### Thus, most captains chose to field after winning the toss. So, in 59 games, captains have chosen to field first, and in just 14 games, the captains have chosen to bat first.

## Top Scorers in IPL 2022

```python
plt.figure(figsize=(16, 6))
sns.countplot(x="top_scorer", data=data)
plt.title("Top Scorers in IPL 2022")
plt.xticks(rotation=-90)
plt.ylabel('Count of matches')
plt.show()
```

 #### Jos Buttler has been a top scorer with 7 matches

## Top Scorers in IPL 2022 with Highscore

```python
figure = px.bar(data, x=data["top_scorer"], 
                y = data["highscore"], 
                color = data["highscore"],
            title="Top Scorers in IPL 2022")
figure.show()
```

#### Jos Buttler has scored 4 centuries, KL Rahul has scored 2 centuries and Rajat Patidar 7 Quinton de Kock has scored 1 century.

## Most Player of the Match Awards

```python
plt.figure(figsize=(16, 5))
sns.countplot(x="player_of_the_match", data=data)
plt.title("Most Player of the Match Awards")
plt.ylabel('Count of matches')
plt.xticks(rotation=-90)
plt.show()
```

#### Kuldeep Yadav is leading in the list of players of the match awards with 4 matches and Jos Buttler with 3 matches.


## Best Bowlers in IPL 2022

```python
plt.figure(figsize=(14, 5))
sns.countplot(x="best_bowling", data=data)
plt.title("Best Bowlers in IPL 2022")
plt.ylabel('Count of matches')
plt.xticks(rotation=-90)
plt.show()
```

#### Yuzvendra Chahal has proved being the best bowler in 5 matches and also Rashid Khan with 4 matches

## Wickets in each venue for both innings

```python
# Set up a figure and axis
plt.figure(figsize=(12, 6))

# Line plot for First Innings Wickets vs Second Innings Wickets
sns.lineplot(x="venue", y="first_ings_wkts", data=data, label='First Innings Wickets', marker='o')
sns.lineplot(x="venue", y="second_ings_wkts", data=data, label='Second Innings Wickets', marker='o')

# Add labels and title
plt.title('Comparison of First Innings Wickets and Second Innings Wickets Across Venues')
plt.xticks(rotation=-90, ha='left')
plt.ylabel('Number of Wickets')
plt.legend()
plt.show()
```

#### So, in the Dr DY Patil Sports Academy in Mumbai and MCA Stadium in Pune, most wickets fall while chasing the target. And in the other 3 stadiums, most wickets fall while setting the target.
